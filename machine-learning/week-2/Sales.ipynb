```python
ratio = 0.8
train_data = new_df[:int(ratio*(len(new_df)))]
test_data = new_df[int(ratio*(len(new_df))):]

print('Train Data Shape:', train_data.shape)
print('Test Data Shape:', test_data.shape)
```

```python
! pip install pmdarima
```

```python
# prompt: perform auto-arima and find best model
import pmdarima as pm

model_auto_ARIMA = pm.auto_arima(train_data, start_p=1, start_q=1,
                      test='adf',       # use adfuller test to find stationarity
                      max_d=3,          # maximum differencing degree
                      max_p=5,          # maximum p degree
                      max_q=5,          # maximum q degree
                      stationary=False,  # set to True if time series is stationary
                      warnings=False,    # disable warnings
                      trace=True,
                      stepwise=True)

print("Auto-ARIMA Model Order:", model_auto_ARIMA.order)
print("Auto-ARIMA Seasonal Order:", model_auto_ARIMA.seasonal_order)
```

```python
# prompt: perform arima

model_auto_ARIMA.fit(train_data)
forecast = model_auto_ARIMA.predict(start=len(train_data), end=len(train_data) + len(test_data) - 1)
forecast = pd.DataFrame(forecast,index = test_data.index,columns=['Prediction'])

# Plot the predictions for validation set
plt.figure(figsize=(20, 6))
plt.plot(train_data, label='Train Data')
plt.plot(test_data, label='Test Data')
plt.plot(forecast, label='Prediction')
plt.legend()
plt.show()
```

```python
# prompt: Evaluate models using metrics like MAE, RMSE, and MAPE.

from sklearn.metrics import mean_absolute_error, mean_squared_error, mean_absolute_percentage_error

# Calculate metrics for the validation set
mae = mean_absolute_error(test_data, forecast)
rmse = mean_squared_error(test_data, forecast, squared=False)
mape = mean_absolute_percentage_error(test_data, forecast)

print("Mean Absolute Error (MAE):", mae)
print("Root Mean Squared Error (RMSE):", rmse)
print("Mean Absolute Percentage Error (MAPE):", mape)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

# Load the data
df = pd.read_csv('Super_Store_data.csv', encoding="latin-1", infer_datetime_format=True)

# Preprocess the data
df['Order Date'] = pd.to_datetime(df['Order Date'])
df = df.sort_values(by='Order Date')
df.drop_duplicates(inplace=True)
df.drop(columns=['Row ID', 'Order ID', 'Product ID', 'Ship Date', 'Ship Mode', 'Customer ID', 'Customer Name', 'Postal Code', 'Product Name', 'City', 'State', 'Country'], axis=1, inplace=True)
df.fillna(0, inplace=True)
```

```python
df
```

```python
df.set_index('Order Date', inplace=True)
```

```python
df.drop(['Category'], axis=1, inplace=True)
```

```python
cat_cols = df.select_dtypes("object").columns

corr_table = df
for col in cat_cols:
    codes, _ = pd.factorize(df[col])
    corr_table[col] = codes

df = corr_table
```

```python
df
```

```python
# prompt: convert to monthly

new_df = df.groupby(pd.Grouper(freq='M')).sum()
new_df
```

```python
new_df.drop(['Segment','Region','Sub-Category'], axis=1, inplace=True)
```

```python
new_df
```

```python
# prompt: plot subplots

# Create subplots
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(15, 5))

# Plot each variable in a separate subplot
for idx, ax in enumerate(axes.flatten()):
    cols = ["Sales", "Quantity", "Discount", "Profit"]
    data = new_df[cols[idx]]

    ax.plot(data, linewidth=1)

    ax.set_title(cols[idx].title())
    ax.xaxis.set_ticks_position("none")
    ax.yaxis.set_ticks_position("none")

# Tight layout
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# prompt: use minmaxscalr

scaler = MinMaxScaler()

# Fit the scaler to the data
scaler.fit(new_df)

# Transform the data
new_df_scaled = scaler.transform(new_df)
```

```python
new_df_scaled
```

```python
features = new_df_scaled
target = new_df_scaled[:,0]
```

```python
# prompt: split

split_ratio = 0.8

train_size = int(split_ratio * len(features))

train_features = features[:train_size]
test_features = features[train_size:]

train_target = target[:train_size]
test_target = target[train_size:]

print("Train features shape:", train_features.shape)
print("Train target shape:", train_target.shape)
print("Test features shape:", test_features.shape)
print("Test target shape:", test_target.shape)
```

```python
win_length = 2
batch_size = 1
num_features = train_features.shape[1]
```

```python
from tensorflow.keras.preprocessing.sequence import TimeseriesGenerator
train_generator = TimeseriesGenerator(train_features, train_target, length=win_length, batch_size=batch_size)
test_generator = TimeseriesGenerator(test_features, test_target, length=win_length, batch_size=batch_size)
```

```python
train_generator[0]
```

```python
model = Sequential()
model.add(LSTM(units=50, input_shape=(win_length, num_features)))
model.add(Dense(1))
#model.compile(optimizer='adam', loss='mse')
```

```python
model.summary()
```

```python
import tensorflow as tf
```

```python
early_stopping = tf.keras.callbacks.EarlyStopping(monitor='loss', patience=5, mode='min')
model.compile(optimizer='adam', loss='mse')
```

```python
history = model.fit(train_generator, epochs=10, validation_data=test_generator, callbacks=[early_stopping], shuffle=False)
```

```python
model.evaluate(test_generator)
```

```python
predictions = model.predict(test_generator)
```

```python
df_final=new_df[predictions.shape[0]*-1:]
```
